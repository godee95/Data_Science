## dataframe_basic_operation

```python
import pandas as pd
from pandas import Series
from pandas import DataFrame

import numpy as np
```

```python
# series addtion operation
s1 = Series(range(1,6), index=list('abced'))
print(s1, type(s1))
s2 = Series(range(5,11), index=list('bcedef'))
print(s2, type(s1))
```

```python
# NaN 값때문에 type이 float형으로 바뀜.
s1 + s2 # 원소별 덧셈을 하게되는데 만약 겹치는 index가 없다면 NaN 으로 출력
```

```python
s1.add(s2)  # s1 +s2랑 같은.. / sub, div, mul도 제공
```

```python
s1.add(s2, fill_value=0) # fill_value는 없는값일경우, 있는 값 그대로 넣기.
```

```python
# dataframe operation
df1 = DataFrame(np.arange(9).reshape(3,3), columns=list('abc'))
print(df1)
df2 = DataFrame(np.arange(16).reshape(4,4), columns=list('abcd'))
print(df2)
```

```python
df1 + df2 # 겹치는게 없다면 NaN
```

```python
df1.add(df2, fill_value=0) # fill_value를 주면 있는 값 그대로.. / fill_value없이하면 NaN
```

```python
#operations for dataframe with series
df = DataFrame(np.arange(16).reshape(4,4), columns=list("abcd"))
print(df)
print()
s = Series(np.arange(10,14), index=list("abcd"))
print(s)
```
