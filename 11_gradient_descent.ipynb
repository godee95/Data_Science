## Simple Gradient Descent
<br>

> - 모델을 학습 시 최적의 매개변수(weight, bias)를 찾는다. <br>
> - 최적인란, 손실 함수(cost function)가 최솟값이 될 때의 매개변수 값이다. (물론 cost가 낮다고 다 좋은 성능을 보여주진 않는다..)<br>
> - 따라서 기울기를 이용하여 손실 함수의 최솟값을 찾으려는 것이 바로 경사 하강법(gradient descent)이다.<br>

$$x_0 = x_0 - \alpha \frac{\partial f}{\partial x_0} $$
<br>
위 수식 처럼 첫 위치에서 기울기를 구하고 기울기가 (+) 이면 음의 방향으로, (-) 이면 양의 방향으로 `α(learning rate)` 만큼 일정 거리만큼 이동한 다음, 그 위치에서 기울기를 다시 구하고 그것을 토대로 다시 일정 거리를 이동하는 것을 반복하여, 순실 함수의 값을 점차적으로 줄여나가는 것이 `gradient descent` 방법이다.
<br>
기존적으로 cost function 이 `convex function`일때 적용이 가능하지만....아닌 경우도 많다는점~_~

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10, 11, 1)
f_x = x ** 2

plt.plot(x, f_x)
plt.show()
```

```python
x_new = 10
derivative = []
y = []
learning_rate = 0.1
for i in range(100):
    old_value = x_new
    derivative.append(old_value - learning_rate * 2 * old_value)
    x_new = old_value - learning_rate * 2 * old_value
    y.append(x_new ** 2)
```

```python
plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```

```python
def sin_function(x):
    return x * np.sin(x ** 2) + 1

def derivitive_f(x):
    return np.sin(x**2) + 2 * (x ** 2) * np.cos(x ** 2)
```

```python
x = np.arange(-3, 3, 0.01)
f_x = sin_function(x)
```

```python
plt.plot(x, f_x)
plt.show()
```

```python
derivitive_f(3)
```

```python
x_new = 1
derivative = []
y = []
learng_rate= 0.01

for i in range(10000):
    old_value = x_new
    x_new = old_value - learng_rate * derivitive_f(old_value)
    derivative.append(x_new)
    y.append(sin_function(x_new))

plt.plot(x, f_x)
plt.scatter(derivative, y)
plt.show()
```
