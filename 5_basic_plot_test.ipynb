## Matplotlib

> Matplotlib는 python에서 data를 차트, plot로 visualization 하는 패키지

> 정형화된 차트나 플롯 이외에도 low api를 사용한 다양한 시각화 기능을 제공<br><br>
        ` 라인 플롯(line plot)`<br>
        ` 스캐터 플롯(scatter plot)`<br>
        ` 컨투어 플롯(contour plot)`<br>
        ` 서피스 플롯(surface plot)`<br>
        ` 박스 플롯(box plot)`<br>
        ` 바 플롯(bar chart)`<br>
        ` 히스토그램(histogram)`
    
matplotlib 시각화 사이트 [클릭](<https://matplotlib.org/gallery.html>)

```python
import numpy as np 
import matplotlib.pyplot as plt

# IPython 에서 제공하는 Rich output 에 대한 표현 방식으로 온라인에서 바로 볼 수 있게 만들어 주는 명령어
%matplotlib inline
```

```python
# rcParams을 이용하여 차트 그림(figure)의 기본 설정을 지정할 수 있다.
# 차트의 크기와 선의 색, 두께등의 기본 값을 설정 가능
plt.rcParams["figure.figsize"] = (10,5)

x = range(100) # x축은 0 ~ 99
y = [value**2 for value in x] # 제곱한 값을 y축으로

plt.plot(x,y) # plot는 line을 그리는 plot이다.
plt.savefig('test.png') # 이미지 파일로 저장
# plt.show()를 안해줘도 디폴트로 있긴하지만 써주는게 좋음.
```

```python
import numpy as np

# 0~99까지의 x축을 기준으로 sin, cos 그래프 그리기
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]
X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

# show 하기전에 plot을 여러개 쓰면 같은 figure에 그려진다.
plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig = plt.figure() # 빈 figure 생성
fig.set_size_inches(10,5) # 크기를 10,5로 설정

ax_1 = fig.add_subplot(1,2,1) # sub로 1행2렬로 만드는데 그중 1열에 들어갈
ax_2 = fig.add_subplot(1,2,2) #2열

ax_1.plot(X_1, Y_1, c="b") # color = blur
ax_2.plot(X_2, Y_2, c="g") # color = green

plt.show();
```

### Set color
- http://matplotlib.org/2.0.2/api/colors_api.html

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value + 100 for value in X_2]

plt.plot(X_1, Y_1, color="#000000") #hex ->block
plt.plot(X_2, Y_2, c="c") #cyan 맑은 청록색
plt.show()
```

### Set linestyle
- https://matplotlib.org/examples/lines_bars_and_markers/linestyles.html
    

```python
plt.plot(X_1, Y_1, c="b", linestyle="dashed") # 점선
plt.plot(X_2, Y_2, c="r", ls="dotted") # 점이 찍힌선

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.title("Two lines") # figure의 title을 지정

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.title('$y = \\frac{ax + b}{test}$') # Latex 타입의 표현으로 수식 표현도 가능하다
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.title('$y = ax+b$')
plt.xlabel('$x_line$') # x축 label 설정
plt.ylabel('y_line') # y축 label 설정
plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")
plt.text(50, 70, "Line_1")

# annotate 그래프에 화살표를 그린후, 그 화살표에 문자열을 출력하는 기능
# plt.annotate(‘문자열',xy,xytext,arrowprops)
plt.annotate('line_2', xy=(50, 100), xytext=(30, 110),
arrowprops=dict(facecolor='black', shrink=0.1)) 
plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')

plt.show()
```

```python
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=False, loc="upper right") # 범례 표시
plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')


plt.show()
```

### Scatter
<br><br>
> 2차원 데이터 즉, 두 개의 실수 데이터 집합의 상관관계를 살펴보는 방법

```python
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x")
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="o")

plt.show()
```

```python
# 데이터가 2차원이 아니라 
# 3차원 혹은 4차원인 경우에는 점 하나의 크기 혹은 색깔을 이용하여 다른 데이터 값을 나타낼 수도 있다.
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2
plt.scatter(x, y, s=area, c=colors, alpha=0.5) # 보통 buble chart 라고 한다
plt.show()
```

### Bar chart (막대 그래프 그리기)
<br><br>
- x 데이터가 카테고리 값인 경우에는 bar 명령과 barh 명령으로 <br>
- 바 차트(bar chart) 시각화를 할 수 있다. <br>
- 가로 방향으로 바 차트를 그리려면 barh 명령을 사용한다.

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]

X = np.arange(0,8,2)
```

```python
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50)
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

```python
data = np.array([[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])


color_list = ['b', 'g', 'r']
data_label = ["A","B","C"]
X = np.arange(data.shape[1])
X
```

```python
data = np.array([[5., 5., 5., 5.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]])

for i in range(3):
   plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0),
           color = color_list[i], label=data_label[i])

plt.legend()
plt.show()

```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)

plt.bar(X, women_pop, color = 'b')
plt.bar(X, men_pop, color = 'r', bottom = 60)

plt.show()
```

```python
women_pop = np.array([5, 30, 45, 22])
men_pop = np.array([5, 25, 50, 20])
X = np.arange(4)
plt.barh(X, women_pop, color = 'r')
plt.barh(X, -men_pop, color = 'b')
plt.show()
```
